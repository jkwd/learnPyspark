```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('date').getOrCreate()
```

```
df = spark.read.csv('FileStore/tables/appl_stock.csv', header=True, inferSchema=True)
```

```
df.printSchema()
```

```
df.show()
```

```
from pyspark.sql.functions import dayofmonth, hour, dayofyear, month, year, weekofyear, format_number, date_format
```

```
df.select(month(df['Date'])).show()
```

```
# df.select(year(df['Date'])).show()
newdf = df.withColumn('year', year(df['Date']))
```

```
results = newdf.groupBy('year').mean().select(['year', 'avg(Close)'])
```

```
results = results.withColumnRenamed('avg(Close)', 'avg close price')
```

```
results.select(['year', format_number('avg close price', 2).alias('avg close price')]).show()
```
