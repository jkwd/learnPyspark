```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('tree_proj').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/dog_food.csv', header=True, inferSchema=True)
```

```
df.printSchema()
```

```
df.head(1)
```

```
from pyspark.ml.classification import DecisionTreeClassifier, RandomForestClassifier
from pyspark.ml.feature import VectorAssembler
```

```
vec_assem = VectorAssembler(inputCols=['A','B','C','D'], outputCol='features')
```

```
vec_df = vec_assem.transform(df)
```

```
rfc = RandomForestClassifier(labelCol='Spoiled', featuresCol='features')
```

```
final_data = vec_df.select(['features', 'Spoiled'])
```

```
model = rfc.fit(final_data)
```
