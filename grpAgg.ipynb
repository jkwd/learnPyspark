```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('agg').getOrCreate()
```

```
df = spark.read.csv('FileStore/tables/sales_info.csv', inferSchema=True, header=True)
```

```
df.printSchema()
```

```
df.show()
```

```
df.groupBy('Company').count().show()
```

```
df.agg({'Sales':'sum'}).show()
```

```
grp_data = df.groupBy('Company')
```

```
grp_data.agg({'Sales':'max'}).show()
```

```
from pyspark.sql.functions import countDistinct, avg, stddev
```

```
df.select(avg('Sales').alias('Avg Sales')).show()
```

```
from pyspark.sql.functions import format_number
```

```
results = df.select(stddev('Sales').alias('std sales'))
```

```
results.select(format_number('std sales', 2).alias('std sales')).show()
```

```
df.show()
```

```
df.orderBy('Sales', ascending=False).show()
```
